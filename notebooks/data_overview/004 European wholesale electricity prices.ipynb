```python
import os
os.chdir('../../')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context='paper', font_scale=2.0, style='whitegrid')
```

```python
electricity_prices_raw = pd.read_csv(
    'data/raw/European wholesale electricity prices.csv'
)

electricity_prices_germany = electricity_prices_raw.loc[electricity_prices_raw['ISO3 Code'] == 'DEU']
electricity_prices_germany.set_index('Date', inplace=True)
electricity_prices_germany.index = pd.to_datetime(electricity_prices_germany.index)
electricity_prices_germany.sample(n=10)
```

```python
fig, ax = plt.subplots(figsize=(14, 4))

electricity_prices_germany['Price (EUR/MWhe)'].plot(ax=ax)
ax.set_title('German Wholesale Electricity Prices in EUR/MWhe')
ax.set_ylabel('Euro/MWhe')
ax.set_xlabel('Date')

# Save plot
fig.savefig("diagrams/data_overview_004_001.svg", format="svg", bbox_inches="tight")
```
