```python
import os
os.chdir('../../')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context='paper', font_scale=1.15, style='whitegrid')
```

```python
storage_levels = pd.read_csv(
    'data/raw/StorageData_GIE_2011-01-01_2023-03-02.csv',
    sep=';',
    index_col='Gas Day Start',
    parse_dates=True,
    decimal='.'
)
```

```python
storage_levels['Gas in storage (TWh)'].sample(n=10).sort_index()
```

```python
fig, ax = plt.subplots(figsize=(14, 4))

storage_levels['Gas in storage (TWh)'].plot(ax=ax)
ax.set_title('German nuatural gas storage levels')
ax.set_ylabel('Gas in storage (TWh)')
ax.set_xlabel('Date')
```

```python
fig, ax = plt.subplots(figsize=(14, 4))

# Filter data to only show data from 2022
storage_levels['Gas in storage (TWh)'].loc['2022'].plot(ax=ax)
ax.set_title('German nuatural gas storage levels 2022')
ax.set_ylabel('Gas in storage (TWh)')
ax.set_xlabel('Date')
```
