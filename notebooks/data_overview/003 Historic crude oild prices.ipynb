```python
import os
os.chdir('../../')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context='paper', font_scale=2.0, style='whitegrid')
```

```python
crude_oil_prices = pd.read_csv(
    'data/raw/Crude oil prices Brent - Europe.csv',
    parse_dates=['DATE'],
    index_col='DATE',
    decimal='.'
)

crude_oil_prices.index.name = 'Date'

# Convert values to float if possible, otherwise to nan
crude_oil_prices = crude_oil_prices.applymap(lambda x: float(x) if x.replace('.', '', 1).isdigit() else float('nan'))

crude_oil_prices.sample(n=10)
```

```python
fig, ax = plt.subplots(figsize=(14, 4))

crude_oil_prices['DCOILBRENTEU'].plot(ax=ax)
ax.set_title('Crude Oil Prices: Brent - Europe')
ax.set_ylabel('USD per Barrel')
ax.set_xlabel('Date')

# Save plot
fig.savefig("diagrams/data_overview_003_001.svg", format="svg", bbox_inches="tight")
```
